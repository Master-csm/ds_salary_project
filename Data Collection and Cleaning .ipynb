```python
df.rename(columns = {'No_of_Reviews':'Reviews','No_of_Stars':'Stars','Queried_Salary':'Salary','Company':'Company_Name','Company_Employees':'Employees', 'Company_Industry':'Industry','Company_Revenue':'Revenue'},inplace=True)
df
```

```python
median = df['Reviews'].median()
median1 = df['Stars'].median()
median2 = df['Date_Since_Posted'].median()
df['Reviews'].fillna(median, inplace = True)
df['Stars'].fillna(median1, inplace = True)
df['Date_Since_Posted'].fillna(median2, inplace = True)
```

```python
df.Stars.fillna(df.Stars.mean()).isnull().any()
```

```python
df.Reviews.fillna(df.Stars.mean()).isnull().any()
```

```python
df.Date_Since_Posted.fillna(df.Date_Since_Posted.mean()).isnull().any()
```
